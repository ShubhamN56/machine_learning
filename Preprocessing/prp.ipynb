```python
cat = list(x.columns[x.dtypes=="object"])
cat
```

```python
con = list(x.columns[x.dtypes!="object"])
con
```

```python
# data cleaning

def replacer(df):
    cat = list(df.columns[df.dtypes == "object"])
    for col in df.columns:
         if col in cat:
            mode = df[col].mode()[0]
            df[col] = df[col].fillna(mode)
         else:
             median = df[col].median()
             df[col] = df[col].fillna(median)

replacer(x)
```

```python
x.to_csv("cars1",index=False)
```

```python
# preprocessing

xcat = x[cat]
xcat
```

```python
xcon = x[con]
xcon
```

```python
#apply ohe in cat

from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(handle_unknown="ignore", 
                    sparse_output=False, 
                    drop="first").set_output(transform="pandas")

xcatpre = ohe.fit_transform(xcat)
xcatpre
```

```python
#con using ss
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().set_output(transform="pandas")

xconpre = scaler.fit_transform(xcon)
xconpre
```

```python
# combiine

xpre = xconpre.join(xcatpre)
xpre.head()
```

```python
#TTS

from sklearn.model_selection import train_test_split

xtr,xts,ytr,yts = train_test_split(xpre,y, test_size=0.2, random_state=21)
```

```python
# model create

from sklearn.linear_model import LinearRegression

m = LinearRegression()

m.fit(xtr,ytr)
```

```python
m.score(xtr,ytr)
```

```python
m.score(xts,yts)
```

```python
# eval

from sklearn.metrics import(
    root_mean_squared_error,
    r2_score,
    mean_absolute_error,
    mean_absolute_percentage_error
)

def eval(m,x,y):
    ypred = m.predict(xts)
    rmse = root_mean_squared_error(y,ypred)
    mae = mean_absolute_error(y,ypred)
    mape = mean_absolute_percentage_error(y,ypred)
    r2 = r2_score(y,ypred)
    
    print(f"rmse = {rmse:.2f}")
    print(f"mae = {mae:.2f}")
    print(f"mape = {mape:.2%}")
    print(f"r2 = {r2:.2%}")
    
eval(m,xts,yts)
```
